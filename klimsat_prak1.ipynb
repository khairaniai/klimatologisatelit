```python
### making NO2 plot using Sentinel-5P OFFL NO2: Offline Nitrogen Dioxide

collection = (ee.ImageCollection('COPERNICUS/S5P/OFFL/L3_NO2') #image collection itu nama data set, beda data set beda nama
  .select('tropospheric_NO2_column_number_density') #band yang tersedia yg data set
  .filterDate('2019-08-01', '2019-08-31'))

band_viz = {
  'min': 0,
  'max': 0.0002, #bisa diliat di dataset utk nilai max min, dan di sini diatur sendiri
  'palette': ['black', 'blue', 'purple', 'cyan', 'green', 'yellow', 'red']
}

map.addLayer(collection.mean(), band_viz, 'S5P N02'); #rata rata dalam rentang waktu yg ditentukan, krn sentinel mungkin ga tiap hari ada jadi nanti dapet brp dlm rentang waktunya dan di rata2in
#Map.setCenter(65.27, 24.11, 4); #untuk memfokuskan peta
map
```

```python
##calculating NDVI using Harmonized Sentinel-2 MSI: MultiSpectral Instrument, Level-2A

band8 = (ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED")
  .select('B8')
  .filterDate('2019-08-01', '2019-08-31'))
band4 = (ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED")
  .select('B4')
  .filterDate('2019-08-01', '2019-08-31'))

NDVI = band8.mean().subtract(band4.mean()).divide(band8.mean().add(band4.mean()))

#Visualize to Map
NDVI_viz = {
  'min': -0.5,
  'max': 1,
  'palette': ['red', 'white', 'green']
}

map.addLayer(NDVI, NDVI_viz, 'NDVI');
map
```
