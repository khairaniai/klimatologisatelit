```python
# @title Default title text
import ee
import geemap

# Initialize the Earth Engine API
ee.Initialize()

# Load the dataset
dataset = ee.ImageCollection('COPERNICUS/DEM/GLO30')

# Select the elevation band
elevation = dataset.select('DEM')

# Define visualization parameters
elevation_vis = {
    'min': 0.0,
    'max': 1000.0,
    'palette': ['0000ff', '00ffff', 'ffff00', 'ff0000', 'ffffff'],
}

# Set the center of the map
Map = geemap.Map(center=[46.529, -6.746], zoom=4)

# Add the elevation layer to the map
Map.addLayer(elevation, elevation_vis, 'DEM')

# Display the map
Map
```
